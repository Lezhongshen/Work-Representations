## Getting and cleaning twitter data

```python
import pandas as pd
from collections import Counter
import tweepy
from AppCred import CONSUMER_KEY, CONSUMER_SECRET
from AppCred import ACCESS_TOKEN, ACCESS_TOKEN_SECRET
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
api = tweepy.API(auth)
```

```python
biden_timeline = api.user_timeline(screen_name = "POTUS")
```

```python
len(biden_timeline)
```

```python
user = api.get_user(screen_name = "POTUS")
statuses_count = user.statuses_count 
print(statuses_count)
```

```python
biden_tweet_list = []

for tweet in tweepy.Cursor(api.user_timeline,screen_name = 'POTUS',tweet_mode = 'extended').items(2271):
    biden_tweet_list.append(tweet)

len(biden_tweet_list)
```

```python
text_list = []

for x in biden_tweet_list:
    text = x._json['full_text']
    text_list.append(text)

text_list_lower = [x.lower() for x in text_list]
#transforming all words into lowercase makes the analysis convenient

text_list_lower
```

```python
like_count_list = []

for x in biden_tweet_list:
    likes = x._json['favorite_count']
    like_count_list.append(likes)
like_count_list
```

```python
retweet_count_list = []

for x in biden_tweet_list:
    retweets = x._json['retweet_count']
    retweet_count_list.append(retweets)
retweet_count_list
```

```python
time_list = []

for x in biden_tweet_list:
    time = x._json['created_at']
    time_list.append(time)
time_list
```

```python
biden_dict = {'text_content': text_list_lower,
             'time': time_list,
             'like_count': like_count_list,
            'retweet_count': retweet_count_list
            }

biden_dataframe = pd.DataFrame(biden_dict)
biden_dataframe
```

```python
#turn tweets created time into date time format

biden_dataframe['time'] = pd.to_datetime(biden_dataframe['time'], format = "%a %b %d %X %z %Y")
biden_dataframe
```

```python
#based on the date to classify the quarter of the text

biden_dataframe.loc[biden_dataframe['time'] < '2021-04-01 00:00:00+00:00','Quarter of year'] = 'Q1'
biden_dataframe.loc[(biden_dataframe['time'] >= '2021-04-01 00:00:00+00:00')&
(biden_dataframe['time'] < '2021-07-01 00:00:00+00:00') ,'Quarter of year'] = 'Q2'
biden_dataframe.loc[(biden_dataframe['time'] >= '2021-07-01 00:00:00+00:00')&
(biden_dataframe['time'] < '2021-10-01 00:00:00+00:00') ,'Quarter of year'] = 'Q3'
biden_dataframe.loc[(biden_dataframe['time'] >= '2021-10-01 00:00:00+00:00')&
(biden_dataframe['time'] < '2022-01-01 00:00:00+00:00') ,'Quarter of year'] = 'Q4'
```

```python
#choose the tweets that were sent last year

biden_dataframe_2021 = biden_dataframe.loc[biden_dataframe['time'] < '2022-01-01 00:00:00+00:00']
biden_dataframe_2021
```

```python
#extract the link, make binary value for whether a text contains a link
#link format: https://t.co/5mjezpnxm0

link_list = []
for x in biden_dataframe_2021['text_content']:
    if 'https://t.co/' in x:
        link_list.append(1)
    else:
        link_list.append(0)
link_list
```

```python
biden_dataframe_2021['link_inculde'] = link_list
biden_dataframe_2021
```

## Word count

Content:

- top 100 word counts for all the text without sorting

- top 100 and last 100 word counts for text sorted by like count

- top 100 and last 100 word counts for text sorted by retweet count


Notes:

- Remove stop words based on stop words file.

- Some tweets are biden's retweets and they have 0 like count. I will remove these tweets' word count.

- Combine all these word counts into one dataframe.

```python
with open ('stop_words.txt','r') as file: 
    stop_words = file.read()
stop_words_list = stop_words.split() 
stop_words_list
```

```python
#word count for all the text, remove stop words

word_join = ' '.join(biden_dataframe_2021.loc[biden_dataframe_2021['like_count'] != 0]['text_content'])
word_all = word_join.split()
#select biden's tweets and exclude biden's rewteets

relevant_words = []
for word in word_all:
    if word not in stop_words_list and word != '—': 
        relevant_words.append(word)
    
tweets_count = Counter(relevant_words)
words_count = tweets_count.most_common()
  
words_count
```

```python
words = [x[0] for x in words_count][0:100]
count = [x[1] for x in words_count][0:100]
```

```python
#sort the dataframe by like counts

biden_dataframe_sort_like = biden_dataframe_2021.sort_values('like_count',ascending = False).copy()
biden_dataframe_sort_like

#0 like means that those are retweets by biden
```

```python
#count words in the top 100 tweets sorted by like amounts

like_join_top100 = ' '.join(biden_dataframe_sort_like.head(100)['text_content'])
like_all_top100 = like_join_top100.split()

relevant_words_top100 = []
for word in like_all_top100:
    if word not in stop_words_list and word != '—': 
        relevant_words_top100.append(word)
    
tweets_count1 = Counter(relevant_words_top100)
words_count1 = tweets_count1.most_common()
  
words_count1
```

```python
#count words in the last 100 tweets sorted by like amounts

like_join_tail100 = ' '.join(biden_dataframe_sort_like.loc[biden_dataframe_sort_like['like_count'] != 0].tail(100)
                             ['text_content'])
like_all_tail100 = like_join_tail100.split()

relevant_words_tail100 = []
for word in like_all_tail100:
    if word not in stop_words_list and word != '—': 
        relevant_words_tail100.append(word)
    
tweets_count2 = Counter(relevant_words_tail100)
words_count2 = tweets_count2.most_common()
  
words_count2
```

```python
#sort the dataframe by retweet counts

biden_dataframe_sort_retweet = biden_dataframe_2021.sort_values('retweet_count',ascending = False).copy()
biden_dataframe_sort_retweet
```

```python
#count words in the top 100 tweets sorted by retweet amounts

retweet_join_top100 = ' '.join(biden_dataframe_sort_retweet.loc[biden_dataframe_sort_retweet['like_count'] != 0]
                               .head(100)['text_content'])
retweet_all_top100 = retweet_join_top100.split()

retweet_words_top100 = []
for word in retweet_all_top100:
    if word not in stop_words_list and word != '—': 
        retweet_words_top100.append(word)
    
tweets_count3 = Counter(retweet_words_top100)
words_count3 = tweets_count3.most_common()
  
words_count3
```

```python
#count words in the last 100 tweets sorted by retweet amounts

retweet_join_tail100 = ' '.join(biden_dataframe_sort_retweet.loc[biden_dataframe_sort_retweet['like_count'] != 0]
                               .tail(100)['text_content'])
retweet_all_tail100 = retweet_join_tail100.split()

retweet_words_tail100 = []
for word in retweet_all_tail100:
    if word not in stop_words_list and word != '—': 
        retweet_words_tail100.append(word)
    
tweets_count4 = Counter(retweet_words_tail100)
words_count4 = tweets_count4.most_common()
  
words_count4
```

```python
#because the length of each count list is different, to make a dataframe, i extract top 100 words in each list.

words_1 = [x[0] for x in words_count1[0:100]]
count_1 = [x[1] for x in words_count1[0:100]]
words_2 = [x[0] for x in words_count2[0:100]]
count_2 = [x[1] for x in words_count2[0:100]]
words_3 = [x[0] for x in words_count3[0:100]]
count_3 = [x[1] for x in words_count3[0:100]]
words_4 = [x[0] for x in words_count4[0:100]]
count_4 = [x[1] for x in words_count4[0:100]]
```

```python
word_count_dict = {'top100_words': words,
             'top100_count': count,
             'like_top100_words': words_1,
             'like_top100_count': count_1,
             'like_tail100_words': words_2,
             'like_tail100_count': count_2,
             'retweet_top100_words': words_3,
             'retweet_top100_count': count_3,
             'retweet_tail100_words': words_4,
             'retweet_tail100_count': count_4
              }

word_count_dataframe = pd.DataFrame(word_count_dict)
word_count_dataframe
```

Based on the word count, we can see that covid related words, infrastructure law related words, economy related words are most mentioned. Climate is also mentioned a lot overall. 

## Add additional binary variables

Based on the observation of Biden's tweets and personal interests, I select some topics to study.

```python
covid_word_list = ['covid-19','vaccinated','vaccine','pandemic','mask','vaccination','social distance']
economic_word_list = ['jobs','economy','economic','unemployment','job','tax']
bipartisan_infrastructure_word_list = ['bipartisan','infrastructure'] #this is a noted law conducted by biden
climate = ['climate']
china = ['china']
```

```python
#make binary value for whether the tweet contains covid-related word

covid_mention = []

for x in biden_dataframe_2021['text_content']:
    if any(word in x for word in covid_word_list):
        covid_mention.append(1)    
    else:
        covid_mention.append(0)
covid_mention
```

```python
len(covid_mention)
```

```python
covid_mention.count(1)
```

```python
#make binary value for whether the tweet contains economic-related word

economy_mention = []

for x in biden_dataframe_2021['text_content']:
    if any(word in x for word in economic_word_list):
        economy_mention.append(1)    
    else:
        economy_mention.append(0)
economy_mention
```

```python
len(economy_mention)
```

```python
economy_mention.count(1)
```

```python
#make binary value for whether the tweet contains infrastructure law-related word

infrastructure_law_mention = []

for x in biden_dataframe_2021['text_content']:
    if any(word in x for word in bipartisan_infrastructure_word_list):
        infrastructure_law_mention.append(1)    
    else:
        infrastructure_law_mention.append(0)
infrastructure_law_mention
```

```python
infrastructure_law_mention.count(1)
```

```python
#make binary value for whether the tweet contains climate

climate_mention = []

for x in biden_dataframe_2021['text_content']:
    if any(word in x for word in climate):
        climate_mention.append(1)    
    else:
        climate_mention.append(0)
climate_mention
```

```python
climate_mention.count(1)
```

```python
#make binary value for whether the tweet contains China

china_mention = []

for x in biden_dataframe_2021['text_content']:
    if any(word in x for word in china):
        china_mention.append(1)    
    else:
        china_mention.append(0)
china_mention
```

```python
china_mention.count(1)
```

```python
biden_dataframe_2021.loc[:,'covid_contain'] = covid_mention
biden_dataframe_2021.loc[:,'economy_mention'] = economy_mention
biden_dataframe_2021.loc[:,'infrastructure_law_mention'] = infrastructure_law_mention
biden_dataframe_2021.loc[:,'climate_mention'] = climate_mention
biden_dataframe_2021.loc[:,'china_mention'] = china_mention

biden_dataframe_2021
```
