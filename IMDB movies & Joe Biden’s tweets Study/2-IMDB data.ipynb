## Getting and cleaning IMDB data

```python
import requests as rq
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
page_num = [51,101,151,201,251,301,351,401,451]

movie_page_list = ['https://www.imdb.com/search/title/?title_type=feature&year=2019-01-01,2019-12-31&ref_=adv_nxt']

for x in page_num:
    link = f'https://www.imdb.com/search/title/?title_type=feature&year=2019-01-01,2019-12-31&start={x}&ref_=adv_nxt'
    movie_page_list.append(link)

movie_page_list
```

```python
# scrape movie name

h3_list = []
movie_name_list = []

for link in movie_page_list:
    movie_imdb = rq.get(link)
    movie_imdb_html = movie_imdb.text
    movie_imdb_soup = BeautifulSoup(movie_imdb_html)
    h3_all = movie_imdb_soup.find_all('h3')
    h3_list.append(h3_all)
    
for x in h3_list:
    for y in x:
        movie_name = y.find('a')
        if movie_name is not None:
            movie_name_list.append(movie_name.text)
        else:
            movie_name_list.append('')
            
movie_name_list = [x for x in movie_name_list if x != '']
movie_name_list
```

```python
len(movie_name_list)
```

```python
# scrape movie rating


rating_10scale_list = []

for link in movie_page_list:
    movie_rate = rq.get(link)
    movie_rate_html = movie_rate.text
    movie_rate_soup = BeautifulSoup(movie_rate_html)
    rating = movie_rate_soup.find_all('div', {'class':'ratings-bar'})
    for x in rating:
        rate = x.find('strong')
        if rate is not None:
            rating_10scale_list.append(rate.text)
        else:
            rating_10scale_list.append('')

rating_10scale_list
# reason why i don't use .find('strong') directly is because one of the movies misses the rate, so i use this way to 
# find out the missing one
```

```python
len(rating_10scale_list)
```

```python
#scrape metascore

metascore_list = []

for link in movie_page_list:
    movie_metascore = rq.get(link)
    movie_metascore_html = movie_metascore.text
    movie_metascore_soup = BeautifulSoup(movie_metascore_html)
    metascore = movie_metascore_soup.find_all('div', {'class':'ratings-bar'})
    for x in metascore:
        score = x.find('div', {'class':'inline-block ratings-metascore'})
        if score is not None:
            metascore_list.append(score.text)
        else:
            metascore_list.append('')

metascore_list = [x.replace('\n','').replace('Metascore','').replace(' ','') for x in metascore_list]
    
metascore_list
```

```python
len(metascore_list)
```

```python
# scrape movie votes 

vote_list = []

for link in movie_page_list:
    movie_vote = rq.get(link)
    movie_vote_html = movie_vote.text
    movie_vote_soup = BeautifulSoup(movie_vote_html)
    vote_elements = movie_vote_soup.find_all('div', {'class':'lister-item-content'})    
    for x in vote_elements:
        vote_elements1 = x.find_all('span', {'name':'nv'})
        if vote_elements1 != []:     
            #there is an empty list in the result
            vote_list.append(vote_elements1[0].text)
        else:
            vote_list.append('')
    
vote_list
```

```python
len(vote_list)
```

```python
# gross in us/canada

gross_list = []

for link in movie_page_list:
    movie_gross = rq.get(link)
    movie_gross_html = movie_gross.text
    movie_gross_soup = BeautifulSoup(movie_gross_html)
    gross_elements = movie_gross_soup.find_all('div', {'class':'lister-item-content'})
    for x in gross_elements:
        gross = x.find_all('span', {'name':'nv'})
        if len(gross) == 2:     #there is an empty list in the result
            gross_list.append(gross[1].text)
        else:
            gross_list.append('')
            
            
gross_list
```

```python
len(gross_list)
```

```python
# time of movies

time_list = []

for link in movie_page_list:
    movie_time = rq.get(link)
    movie_time_html = movie_time.text
    movie_time_soup = BeautifulSoup(movie_time_html)
    time_elements = movie_time_soup.find_all('p', {'class':'text-muted'})
    for x in time_elements:
        time = x.find('span', {'class':'runtime'})
        if time is not None:
            time_list.append(time.text)
        else:
            time_list.append('')

time_list = [x for x in time_list if x != '']   
#there are two 'class: text-muted and the second one has no 'class: runtime'    
time_list
```

```python
len(time_list)
```

```python
# genre list

genre_list = []

for link in movie_page_list:
    movie_genre = rq.get(link)
    movie_genre_html = movie_genre.text
    movie_genre_soup = BeautifulSoup(movie_genre_html)
    genre_elements = movie_genre_soup.find_all('p', {'class':'text-muted'})
    for x in genre_elements:
        genre = x.find('span', {'class':'genre'})
        if genre is not None:
            genre_list.append(genre.text)
        else:
            genre_list.append('')

genre_list = [x for x in genre_list if x != '']
genre_list = [x.strip() for x in genre_list]
#there are two 'class: text-muted and the second one has no 'class: runtime'    
genre_list
```

```python
len(genre_list)
```

```python
#movie detailed information page

h3_list = []
movie_link_list = []

for link in movie_page_list:
    movie_imdb = rq.get(link)
    movie_imdb_html = movie_imdb.text
    movie_imdb_soup = BeautifulSoup(movie_imdb_html)
    h3_all = movie_imdb_soup.find_all('h3')
    h3_list.append(h3_all)
    
for x in h3_list:
    for y in x:
        movie_name = y.find('a')
        if movie_name is not None:
            movie_name = str(movie_name)
            movie_link_list.append(re.findall(r'href="[\w\d/_]*"',movie_name))
        else:
            movie_link_list.append('')
    # there is a h3 called 'recently review' which is not the movies name so has no 'a' and should be moved.


movie_link_list = [str(x) for x in movie_link_list]
movie_link_list = [x.replace('href="','https://www.imdb.com') for x in movie_link_list]
movie_link_list = [x.replace('"','') for x in movie_link_list]
movie_link_list = [x.replace("['",'').replace("']",'') for x in movie_link_list]
movie_link_list = [x for x in movie_link_list if x != '']

movie_link_list
```

```python
#scrape budget, opening week box office and world-wide box office 

budget_list = []
open_week_list = []
gross_world_list = []

for x in movie_link_list:
    movie_budget = rq.get(x)
    movie_budget_html = movie_budget.text
    movie_budget_soup = BeautifulSoup(movie_budget_html)
    box_office = movie_budget_soup.find('section', {'data-testid':'BoxOffice'})
    if box_office is not None:
        budget_elements = box_office.find('li',{'data-testid':'title-boxoffice-budget'})
        if budget_elements is not None:
            budget = budget_elements.find('span',{'class':'ipc-metadata-list-item__list-content-item'})
            budget_list.append(budget.text)
        else:
            budget_list.append('')
    else:
        budget_list.append('')

               
for x in movie_link_list:
    movie_open_week = rq.get(x)
    movie_open_week_html = movie_open_week.text
    movie_open_week_soup = BeautifulSoup(movie_open_week_html)
    box_office_1 = movie_open_week_soup.find('section', {'data-testid':'BoxOffice'})
    if box_office_1 is not None:
        opening_elements = box_office_1.find('li',{'data-testid':'title-boxoffice-openingweekenddomestic'})
        if opening_elements is not None:
            opening = opening_elements.find('span',{'class':'ipc-metadata-list-item__list-content-item'})
            open_week_list.append(opening.text)
        else:
            open_week_list.append('')
    else:
        open_week_list.append('')
        

for x in movie_link_list:
    movie_gross_world = rq.get(x)
    movie_gross_world_html = movie_gross_world.text
    movie_gross_world_soup = BeautifulSoup(movie_gross_world_html)
    box_office_2 = movie_gross_world_soup.find('section', {'data-testid':'BoxOffice'})
    if box_office_2 is not None:
        gross_world_elements = box_office_2.find('li',{'data-testid':'title-boxoffice-cumulativeworldwidegross'})
        if gross_world_elements is not None:
            gross_world = gross_world_elements.find('span',{'class':'ipc-metadata-list-item__list-content-item'})
            gross_world_list.append(gross_world.text)
        else:
            gross_world_list.append('')
    else:
        gross_world_list.append('')      
        

        
```

```python
#clean the budgets, turn them into the same unit US dollars

budget_list_clean = []

for x in budget_list:
    a = x.replace('(estimated)','').replace(' ','')
    if a != '':
        if a[0] == '$':
            b = a.replace('$','').replace(',','')
            budget_list_clean.append(float(b)/1000000)
        elif 'DKK' in a:
            c = a.replace('DKK','').replace(',','').replace(' ','')
            budget_list_clean.append(float(c)/6.67/1000000)
        elif '€' in a:
            d = a.replace('€','').replace(',','')
            budget_list_clean.append(float(d)/0.893/1000000)
        elif a[0] == 'A':     #there are A$ and CA$ in the list, so don't use 'A$' in a
            e = a.replace('A$','').replace(',','')
            budget_list_clean.append(float(e)/1.439/1000000)
        elif '£' in a:
            f = a.replace('£','').replace(',','')
            budget_list_clean.append(float(f)/0.784/1000000)
        elif 'SEK' in a:
            g = a.replace('SEK','').replace(',','').replace(' ','')
            budget_list_clean.append(float(g)/9.457/1000000)
        elif '₹' in a:
            h = a.replace('₹','').replace(',','')
            budget_list_clean.append(float(h)/70.394/1000000)
        elif 'CZK' in a:
            i = a.replace('CZK','').replace(',','').replace(' ','')
            budget_list_clean.append(float(i)/22.9/1000000)
        elif 'CA$' in a:
            j = a.replace('CA$','').replace(',','')
            budget_list_clean.append(float(j)/1.327/1000000)
        elif 'NOK' in a:
            k = a.replace('NOK','').replace(',','').replace(' ','')
            budget_list_clean.append(float(k)/8.802/1000000)
        elif 'RUR' in a:
            m = a.replace('RUR','').replace(',','').replace(' ','')
            budget_list_clean.append(float(m)/64.687/1000000)
    else:
        budget_list_clean.append(a)
budget_list_clean
```
